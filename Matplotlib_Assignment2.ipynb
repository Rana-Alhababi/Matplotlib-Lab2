# Q4: Display multiple line charts in one row using the charts given below.
- Give each chart a title
- Add a super title 'my data visulaization assignment'.

```python
#super title = my data visulaization assignment

#title = First Chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

#title = Second Chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

#title = Third Chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])
```

```python
# write your code here ^_^

# Create a figure and subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot the first chart
axs[0].plot(x_plot1, y_plot1, 'o-', linewidth=2, markersize=6)
axs[0].set_title('First Chart')

# Plot the second chart
axs[1].plot(x_plot2, y_plot2, 'o-', linewidth=2, markersize=6)
axs[1].set_title('Second Chart')

# Plot the third chart
axs[2].plot(x_plot3, y_plot3, 'o-', linewidth=2, markersize=6)
axs[2].set_title('Third Chart')

# Set the super title
fig.suptitle('My Data Visualization Assignment', fontsize=16)

# Adjust the spacing between subplots
plt.tight_layout()

# Display the charts
plt.show()
```

# Q5: Define an array that has a values of your top three TV series and define another array to hold the rating of the three TV series out of 10. Create a bar chart using the two arrays
- Set a new color for the bars.
- Change the width of the bars to width 0.5.

```python
# write your code here ^_^
series = np.array(['Gossip Girl', 'Breaking Bad', 'Friends'])
ratings = np.array([8, 6, 4])

# Set a new color for the bars
color = 'purple'

# Change the width of the bars to 0.5
width = 0.5

# Create the bar chart
plt.bar(series, ratings, color=color, width=width)

# Add labels and title
plt.xlabel('TV Series')
plt.ylabel('Rating (out of 10)')
plt.title('Top TV Series Ratings')

# Display the chart
plt.show()
```

# Q6: Create a scatter plot  using two arrays. The first array represents carAge and the second array represents carSpeed, set a specific color for each dot by using an array of different colors and add labels for the plot
| carAge | carSpeed |
| ----- | -------- |
| 5    | 99       |
| 7    | 86       |
| 8    |87      |
| 7    | 88       |
| 2    | 111       |
| 17    | 86       |
| 2    | 103      |
| 9    | 87       |
| 4    | 94       |
| 11    | 78      |

- Set a new color for each dot by using an array of different colors.
- Change the size of the dots to size 350.
- Add labels for x-axis and y-axis.
- Adjust the transparency of the dots with the alpha argument. let the alpha value be 0.4

```python
# write your code here ^_^

carAge = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11])
carSpeed = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78])

# Define an array of different colors for each dot
colors = ['red', 'green', 'blue', 'orange', 'purple', 'yellow', 'cyan', 'magenta', 'brown', 'pink']

# Set the size of the dots to 350
dot_size = 350

# Set the transparency of the dots to 0.4
dot_alpha = 0.4

# Create the scatter plot
plt.scatter(carAge, carSpeed, c=colors, s=dot_size, alpha=dot_alpha)

# Add labels for x-axis and y-axis
plt.xlabel('Car Age')
plt.ylabel('Car Speed')

# Display the plot
plt.show()
```

# Q8: Create a histogram with 250 values and 15 bars

```python
# write your code here ^_^

# Generate 250 random values
values = np.random.randn(250)

# Set the number of bins/bar to 15
num_bins = 15

# Set the color of the histogram to light blue
color = 'lightblue'
# Create the histogram
plt.hist(values, bins=num_bins ,color=color)

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

# Display the histogram
plt.show()
```
